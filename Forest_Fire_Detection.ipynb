```python
import cv2
import vonage
from google.colab.patches import cv2_imshow
from ultralytics import YOLO


model = YOLO("/content/gdrive/My Drive/Fire/detect/train/weights/best.pt")



client = Vonage(Auth(api_key=VONAGE_API_KEY, api_secret=VONAGE_API_SECRET))

VONAGE_API_KEY = 'ENTER_API_KEY'
VONAGE_API_SECRET = 'ENTER_API_SECRET'
TO_NUMBER = 'ENTER_TO_NUMBER'
FROM_NUMBER = 'ENTER_FROM_NUMBER'

def send_sms():
  responseData = SmsMessage(
      to=TO_NUMBER,
      from_=FROM_NUMBER,
      text="ðŸ”¥ Fire detected! Please take immediate action.",
      )
  response: SmsResponse = client.sms.send(responseData)
  print(response)


video_path = "ENTER_VIDEO_PATH"
cap = cv2.VideoCapture(video_path)


frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))


output_path = "ENTER_OUTPUT_PATH"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))


class_names = {0: "Fire", 1: "Smoke"}
colors = {0: (255, 165, 0), 1: (192, 192,192 )}

alert_sent = False


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0]
            class_id = int(box.cls[0])

            label = f"{class_names[class_id]}: {confidence:.2f}"
            color = colors.get(class_id, (255, 255, 255))

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)


            if class_id == 0 and not alert_sent:
                send_sms()
                alert_sent = True

    out.write(frame)
    cv2_imshow(frame)


cap.release()
out.release()

print(f"ðŸ”¥ Fire-detected video saved at: {output_path}")
```
